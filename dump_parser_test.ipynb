```python
def clean_file(filename, number):
    df = pd.read_csv('Dump/{0}'.format(filename), header=None)
    new_df = df[df[1]!='[]'].reset_index(drop=True)
    new_df.to_csv('Dump/cleaned_files/links_{0}_cleaned.csv'.format(number), index=False, header=None)
```

```python
for file in os.listdir('Dump'):
    if file.startswith('links_'):
        number = file[file.find('_') + 1:file.find('.')]
        print(file, number)
        clean_file(file, number)
```

```python
for filename in os.listdir('Dump/cleaned_files'):
    links = defaultdict()
    try:
        # print(filename)
        with open('Dump/cleaned_files/{0}'.format(filename), encoding='UTF-8') as f:
            reader=csv.reader(f, skipinitialspace=True, quotechar="\"")
            for row in reader:
                # print(row[0], row[1])
                links[row[0]]=row[1]
    except:
        print('error with file {0}'.format(filename))
        continue
    with open('Dump/cleaned_files/all_links.csv','a', encoding='UTF-8') as f:
        w = csv.writer(f)
        w.writerows(links.items())
```
